```python
import pandas as pd
```

```python
file_path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv'
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(file_path, names= headers)
df.head()
```

```python
df.size
df.shape
df.info()
df.count()
```

```python
import numpy as np
df.replace('?', np.NaN, inplace=True)
df.head()
```

#### Dealing with missing data

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts())
```

```python
avg_mean = df['normalized-losses'].astype('float').mean(axis=0)
avg_mean
```

```python
avg_bore = df['bore'].astype('float').mean(axis=0)
avg_bore
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
avg_stroke
```

```python
avg_comp = df['compression-ratio'].astype('float').mean(axis=0)
avg_comp
```

```python
avg_hp =df['horsepower'].astype('float').mean(axis=0)
avg_hp
```

```python
avg_rpm = df['peak-rpm'].astype('float').mean(axis=0)
avg_rpm
```

```python
nod_max= df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.NaN, nod_max, inplace=True )
```

```python
df['normalized-losses'].replace(np.NaN, avg_mean, inplace=True)
```

```python
df['bore'].replace(np.NaN, avg_bore, inplace=True)
```

```python
df['stroke'].replace(np.NaN, avg_stroke, inplace=True)
```

```python
df['compression-ratio'].replace(np.NaN, avg_comp, inplace=True)
```

```python
df['horsepower'].replace(np.NaN, avg_hp, inplace=True)
```

```python
df['peak-rpm'].replace(np.NaN, avg_rpm, inplace=True)
```

```python
df.dropna(subset=['price'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts())
```

```python
df.head()
```

#### Checking for correct data types

```python
df.dtypes
```

```python
df[['normalized-losses', 'horsepower']] = df[['normalized-losses', 'horsepower']].astype('int')
df[['peak-rpm']] = df[['peak-rpm']].astype('int')
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype('float')
df[['price']] = df[['price']].astype('int')
```

```python
df.dtypes
```

```python
df.head()
```

#### Data Normalization

```python
df['height'] = df['height'] / df['height'] .max()
df['height']
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df[['length', 'width', 'height']].head(n=5)
```

#### Dummy Variable creation

```python
dummy_var_ft = pd.get_dummies(df['fuel-type'])
dummy_var_ft.head()
```

```python
dummy_var_ft.rename(columns={'diesel':'fuel-type-diesel', 'gas': 'fuel-type-gas'}, inplace=True)
df= pd.concat([df, dummy_var_ft], axis=1)
```

```python
df.head()
```

```python
df.drop('fuel-type', axis=1, inplace=True)
```

```python
df.head()
```

```python
dummy_var_as = pd.get_dummies(df['aspiration'])
dummy_var_as.head()
dummy_var_as.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
df= pd.concat([df, dummy_var_as], axis=1)
df.drop('aspiration', axis=1, inplace=True)
df.head()
```

```python
df.to_csv('car_values.csv')
```

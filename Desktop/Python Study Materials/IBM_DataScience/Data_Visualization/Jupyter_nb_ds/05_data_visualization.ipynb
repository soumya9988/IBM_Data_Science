## Data Visualisation using matplotlib

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
india_dataset = df_can.loc['India', years]
india_dataset.plot(kind='line')
plt.title('Immigration from India')
plt.xlabel('Years')
plt.ylabel('No. of immigrants')
plt.show()
```

#### Plot comparing two values

```python
df_can_1 = df_can.loc[['United Kingdom of Great Britain and Northern Ireland', 'Germany'], years]
df_can_2 = df_can_1.transpose()
df_can_2.head()
```

```python
df_can_2.plot(kind='line')
```

```python
df_can.sort_values(by='Total', ascending=False, inplace=True, axis=0)
df_can_sort = df_can.iloc[0:5, 3:-1]
df_can_transpose = df_can_sort.transpose()

df_can_transpose.plot(kind='line', figsize=(15,10))
plt.title('Top 5 Immigrant Country to Canada')
plt.xlabel('Year')
plt.ylabel('No. of immigrants')
plt.show()
```

### Area plots

```python
df_can_transpose.plot(kind='area', 
                      figsize=(15,10),
                      stacked=False,
                      alpha = 0.25)
plt.title('Top 5 Immigrant Country to Canada')
plt.xlabel('Year')
plt.ylabel('No. of immigrants')
plt.show()
```

## Histograms

#### A histogram is a way of representing the frequency distribution of a numeric dataset. 

```python
df_can['2000'].plot(kind='hist', figsize=(15,10))
plt.title('Immigration to Canada')
plt.xlabel('No.of immigrants')
plt.ylabel('No. of Countries')
plt.show()
```

## Bar chart

#### Unlike a histogram, a bar chart also known as a bar graph is a type of plot where the length of each bar is proportional to the value of the item that it represents.

```python
df_can_i = df_can.loc['Iceland', years]
df_can_i.plot(kind='bar')
plt.title('Immigration trend of Iceland to Canada')
plt.xlabel('Year')
plt.ylabel('No. of immigrants')
plt.show()
```
